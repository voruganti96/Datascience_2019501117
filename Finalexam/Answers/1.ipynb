import pandas as pd
bse = pd.read_csv("BSE_Sensex_Index.csv")
#bse["Date"]=pd.to_numeric(bse["Date"])
bse['Date'] = pd.to_numeric(bse.Date.str.replace('/',''))
bse['ddate'] =bse['Date'].diff(1)
bse['dopen'] = bse['Open'].diff(1)
bse['dhigh'] = bse['High'].diff(1)
bse['dlow'] = bse['Low'].diff(1)
bse['dclose'] = bse['Close'].diff(1)
bse['dvol'] = bse['Volume'].diff(1)
bse['dAdjc'] = bse['Adj Close'].diff(1)
bse

```python
bse_mean=bse[["dopen","dhigh","dlow","dclose","dvol","dAdjc"]].mean()
print("Mean of each new columns ")
print(bse_mean)
bse_max=bse[["dopen","dhigh","dlow","dclose","dvol","dAdjc"]].max()
print("Max of each new columns ")
print(bse_max)
bse_var=bse[["dopen","dhigh","dlow","dclose","dvol","dAdjc"]].var()
print("Var of each new columns ")
print(bse_var)
bse_quartile=bse[["dopen","dhigh","dlow","dclose","dvol","dAdjc"]].quantile([0.25,0.5,0.75])
print("Respective quartile of each new columns ")
print(bse_quartile)
```
